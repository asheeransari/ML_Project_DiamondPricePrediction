```python
cut_val = list(data['cut'].unique())
color_val = list(data['color'].unique())
clarity_val = list(data['clarity'].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordEnc = OrdinalEncoder(categories = [cut_val,clarity_val,color_val])
```

```python
cut_val = ['Ideal','Premium', 'Very Good', 'Good', 'Fair']
```

```python
clarity_val = ['I1','SI2','SI1','VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
```

```python
color_val = ['D','E','F', 'G', 'H', 'I','J']
```

```python
new_data = ordEnc.fit_transform(data[['cut','clarity','color']])
```

```python
new_data = pd.DataFrame(new_data,columns=ordEnc.get_feature_names_out())
new_data.head()
```

```python
data = pd.concat([data,new_data],axis = 1)
```
